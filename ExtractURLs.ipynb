```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://canberra.craigslist.org/"
```

```python
response = requests.get(url)
```

```python
response
```

```python
data = response.text
```

```python
data
```

```python
soup = BeautifulSoup(data, "html.parser")
```

```python
tags = soup.find_all('a')
```

```python
for tag in tags:
    print(tag.get('href'))
```
